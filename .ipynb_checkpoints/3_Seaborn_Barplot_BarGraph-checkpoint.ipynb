# Draw Seaborn Barplot / BarGraph

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
tips_df = sns.load_dataset("tips")
tips_df
```

```python
sns.barplot()
```

```python
sns.barplot(x = tips_df.total_bill)

plt.show()
```

```python
sns.barplot(x = tips_df.day, y = tips_df.total_bill)

plt.show()
```

```python
sns.barplot(x = tips_df.day, y = tips_df.total_bill, hue = tips_df.sex)

plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', 
           data = tips_df)

plt.show()
```

```python
order =  ['Sun', 'Thur', 'Fri', 'Sat']

sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', 
           data = tips_df, order = order)

plt.show()
```

```python
hue_order = ["Female", "Male"]

sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', 
           data = tips_df, hue_order = hue_order)

plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', 
           data = tips_df, estimator = np.mean)

plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', 
           data = tips_df, estimator = np.max)

plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', 
           data = tips_df, ci = 20) # ci mean confidence interval

plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', 
           data = tips_df, estimator = np.max, n_boot = 2)

plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', 
           data = tips_df, orient = 'v')

plt.show()
```

```python
sns.barplot(y = 'day', x = 'total_bill', hue = 'sex', 
           data = tips_df)

plt.show()
```

```python
sns.barplot(x = 'total_bill', y = 'size', hue = 'sex', 
            data = tips_df, orient = 'h')

plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', 
           data = tips_df, color = 'g')

plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', 
           data = tips_df, palette = "magma")

plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', 
           data = tips_df, saturation = 0.4)

plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', 
           data = tips_df, errcolor = '0.5')

plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', 
           data = tips_df, errwidth = 11)

plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', 
           data = tips_df, capsize = 1)

plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', 
           data = tips_df, dodge = True)

plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', 
           data = tips_df, dodge = False)

plt.show()
```

```python
kwargs = {'alpha': 0.9, 'linestyle': ':', 'linewidth': 5, 'edgecolor': 'k'}
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', 
           data = tips_df, **kwargs)

plt.show()
```

```python
sns.set()
kwargs = {'alpha': 0.9, 'linestyle': ':', 'linewidth': 5, 'edgecolor': 'k'}
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', 
           data = tips_df, **kwargs)

plt.show()
```

```python
sns.set()
kwargs = {'alpha': 0.9, 'linestyle': ':', 'linewidth': 5, 'edgecolor': 'k'}
sns.barplot(x = 'day', y = 'total_bill', 
           data = tips_df, **kwargs)

plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', 
           data = tips_df, alpha = 0.9, linestyle = "-.", linewidth = 6, 
           edgecolor = "g")

plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', 
           data = tips_df, alpha = 0.9, linestyle = "-.", linewidth = 3, 
           edgecolor = "g")

plt.show()
```

```python
ax = sns.barplot(x = 'day', y = 'total_bill', 
           data = tips_df, alpha = 0.9, linestyle = "-.", linewidth = 3, 
           edgecolor = "g")

ax.set(title = "Barplot Of Tips DataFrame", 
      xlabel = "Days", 
      ylabel = "Total Bill")
plt.show()
```

```python
plt.figure(figsize = (16, 9))

sns.barplot(x = 'day', y = 'total_bill', 
           data = tips_df, alpha = 1, linestyle = "-.", linewidth = 3, 
           edgecolor = "k")

plt.title("Batplot Of Days And Total Bill", fontsize = 20)
plt.xlabel("Days", fontsize = 15)
plt.ylabel("Total Bill", fontsize = 15)

plt.savefig("Batplot Of Days And Total Bill")

plt.show()
```
