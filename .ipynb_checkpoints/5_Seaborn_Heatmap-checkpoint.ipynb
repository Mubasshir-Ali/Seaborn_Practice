# Draw Seaborn Heatmap

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr_2d = np.linspace(1, 5, 12).reshape(4, 3)
arr_2d
```

```python
sns.heatmap(arr_2d)

plt.show()
```

```python
globalwarming_df = pd.read_csv("./seaborn_data/Who_is_responsible_for_global_warming.csv")

globalwarming_df
```

```python
globalwarming_df = globalwarming_df.drop(columns = ["Country Code", "Indicator Name", "Indicator Code"], axis = 1).set_index("Country Name")
globalwarming_df.head(6)
```

```python
sns.heatmap(globalwarming_df)

plt.show()
```

```python
plt.figure(figsize = (16, 9))
sns.heatmap(globalwarming_df)

plt.show()
```

```python
plt.figure(figsize = (16, 9))
sns.heatmap(globalwarming_df, vmin = 0, vmax = 21)

plt.show()
```

```python
plt.figure(figsize = (16, 9))
sns.heatmap(globalwarming_df, cmap = "coolwarm")

plt.show()
```

```python
plt.figure(figsize = (16, 9))
sns.heatmap(globalwarming_df, cmap = "coolwarm", annot = True)

plt.show()
```

```python
arr_2d
```

```python
annot_arr = np.array([['a00', 'a01', 'a02'], ['a10', 'a11', 'a12'], ['a20', 'a21', 'a22'], ['a30', 'a31', 'a32']])
annot_arr
```

```python
sns.heatmap(arr_2d, annot = annot_arr, fmt = "s")

plt.show()
```

```python
plt.figure(figsize = (16, 9))
annot_kws = {"fontsize": 10, 
            "fontstyle": "italic", 
            "color": "k", 
            "alpha": 0.9, 
            "rotation": "vertical", 
            "verticalalignment": "center", 
            "backgroundcolor": "w"}
sns.heatmap(globalwarming_df, cmap = "coolwarm", annot = True, annot_kws = annot_kws)

plt.show()
```

```python
plt.figure(figsize = (16, 9))
sns.heatmap(globalwarming_df, cmap = "coolwarm", annot = True, linewidth = 4, linecolor = 'k')

plt.show()
```

```python
plt.figure(figsize = (16, 9))
sns.heatmap(globalwarming_df, cbar = False)

plt.show()
```

```python
plt.figure(figsize = (16, 9))
sns.heatmap(globalwarming_df, cbar = False, xticklabels = False, yticklabels = False)

plt.show()
```

```python
plt.figure(figsize = (16, 9))
cbar_kws = {"orientation": "horizontal", 
           "shrink": 1, 
           "extend": "min", 
           "extendfrac": 0.1, 
           "ticks": np.arange(0, 22), 
           "drawedges": True}
sns.heatmap(globalwarming_df, cbar_kws = cbar_kws)

plt.show()
```

```python
plt.figure(figsize = (16, 9))

sns.heatmap(globalwarming_df, xticklabels = np.arange(0, 15), yticklabels = False)

plt.show()
```

```python
plt.figure(figsize = (16, 9))
country_code = ['USA', 'GBR', 'IND', 'CHN', 'RUS', 'AUS', 'FRA', 'DEU', 'CAN', 'BRA', 'ARG', 'PAK', 'NPL', 'BGD', 'JPN']
sns.heatmap(globalwarming_df, xticklabels = np.arange(0, 15), yticklabels = country_code)

plt.show()
```

```python
plt.figure(figsize = (16, 9))
ax = sns.heatmap(globalwarming_df)
ax.set(title = "Heatmap Of 'Who Is Responsible For Global Warming'", 
      xlabel = "C02 Emissions (metric tons per capita) Per Year", 
      ylabel = "Country Name")

plt.show()
```

```python
plt.figure(figsize = (16, 9))
ax = sns.heatmap(globalwarming_df)
ax.set(title = "Heatmap Of 'Who Is Responsible For Global Warming'", 
      xlabel = "C02 Emissions (metric tons per capita) Per Year", 
      ylabel = "Country Name")

sns.set(font_scale = 5)
plt.show()
```

```python
globalwarming_df.corr()
```

```python
plt.figure(figsize = (16, 9))
ax = sns.heatmap(globalwarming_df.corr(), annot = True, linewidth = 3)
ax.tick_params(size = 10, color = 'w', labelsize = 10, labelcolor = 'w')
plt.title("Heatmap Of 'Who is responsible for global warming'", fontsize = 25)

plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
cancer_dataset
```

```python
cancer_df = pd.DataFrame(np.c_[cancer_dataset['data'], cancer_dataset['target']], 
                        columns = np.append(cancer_dataset['feature_names'], ['target']))

cancer_df
```

```python
plt.figure(figsize = (25, 25))
sns.heatmap(cancer_df.corr(), annot = True, linewidth = 3)
ax.tick_params(size = 10, color = 'w', labelsize = 10, labelcolor = 'w')
plt.title("Heatmap Of 'Beast Cancer Patients'", fontsize = 25)

plt.show()
```
