```python
plt.figure(figsize = (16, 9))
sns.set()
sns.distplot(tips_df["total_bill"], 
             hist_kws = {'color': '#DC143C', 'edgecolor': '#aaff00', 
                        'linewidth':5, 'linestyle':'--', 'alpha':0.9}, 
            
            kde = False,
            fit = norm,
            fit_kws = {'color': '#8e00ce', 
                        'linewidth':11, 'linestyle':'--', 'alpha':0.4},
            rug = True, 
            rug_kws = {'color': '#0426d0', 'edgecolor': '#00dbff', 
                        'linewidth':5, 'linestyle':'--', 'alpha':0.9})

plt.show()
```

```python
plt.figure(figsize = (16, 9))
sns.set()
sns.distplot(tips_df["total_bill"], 
             hist_kws = {'color': '#DC143C', 'edgecolor': '#aaff00', 
                        'linewidth':5, 'linestyle':'--', 'alpha':0.9}, 
            
            kde = False,
            fit = norm,
            fit_kws = {'color': '#8e00ce', 
                        'linewidth':11, 'linestyle':'--', 'alpha':0.4},
            rug = True, 
            rug_kws = {'color': '#0426d0', 'edgecolor': '#00dbff', 
                        'linewidth':5, 'linestyle':'--', 'alpha':0.9}, 
            label = "TB")

plt.title("Histogram Of Restaurent Total Bill", fontsize = 20)
plt.xlabel("Total Bill", fontsize = 15)
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (16, 9))
sns.distplot(tips_df["total_bill"], bins = 9, label = "total_bill")
sns.distplot(tips_df["tip"], bins = 9, label = "tip")
sns.distplot(tips_df["size"], bins = 9, label = "size")

plt.legend()
plt.show()
```
