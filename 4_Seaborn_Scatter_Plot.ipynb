# Draw Seaborn Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
titanic_df = sns.load_dataset("titanic")
titanic_df
```

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df)

plt.show()
```

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, 
               hue = "sex")

plt.show()
```

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, 
               hue = "sex", style = "who")

plt.show()
```

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, 
               hue = "sex", style = "who", size = "who")

plt.show()
```

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, 
               hue = "sex", style = "who", size = "who", 
               sizes = (100, 400))

plt.show()
```

```python
plt.figure(figsize = (19, 9))
sns.scatterplot(x = "age", y = "fare", data = titanic_df, 
               hue = "sex", style = "who", size = "who", 
               sizes = (100, 400))

plt.show()
```

```python
plt.figure(figsize = (19, 9))
sns.scatterplot(x = "who", y = "fare", data = titanic_df, 
               hue = "alive", style = "alive", size = "who", 
               sizes = (100, 400))

plt.show()
```

```python
plt.figure(figsize = (19, 9))
sns.scatterplot(x = "who", y = "fare", data = titanic_df, 
               hue = "alive", style = "alive", size = "who", 
               sizes = (100, 400), palette = "inferno_r", alpha = .7)

plt.show()
```

```python
plt.figure(figsize = (19, 9))
sns.scatterplot(x = "age", y = "fare", data = titanic_df)
sns.lineplot(x = "age", y = "fare", data = titanic_df)
sns.barplot(x = "age", y = "fare", data = titanic_df)

plt.show()
```
